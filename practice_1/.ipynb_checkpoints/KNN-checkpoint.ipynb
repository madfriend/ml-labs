```python
%matplotlib inline
data = []
with open("chips.txt") as dataset:
    for line in dataset:
        l = line.strip().split(',')
        l[-1] = int(l[-1])
        l[:2] = map(float, l[:2])
        data.append(l)
print data[:5]
```

```python
import matplotlib.pyplot as plt
colors = ['red', 'blue']
for X, Y, cls in data:
    plt.scatter(X, Y, c=colors[cls], label=colors[cls], edgecolors='none')
plt.grid(True)
plt.show()
    
```

```python
from sklearn import cross_validation
X_train, X_test, Y_train, Y_test = cross_validation.train_test_split(
    data, test_size=0.25, random_state=0)
print X_test
```
